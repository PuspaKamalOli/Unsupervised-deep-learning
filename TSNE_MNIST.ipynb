```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import plotly.express as px
(x_train, y_train), (_, _) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
# Reshape and normalize data
x_train = x_train.reshape((x_train.shape[0], -1)) / 255.0
```

```python
# Subset of the data (for faster computation)
subset_size = 10000
x_subset = x_train[:subset_size]
y_subset = y_train[:subset_size]
```

```python
#model and training
tsne = TSNE(n_components=2, random_state=42)
x_tsne = tsne.fit_transform(x_subset)
```

```python
class_labels = {
    0: 'T-shirt/top',
    1: 'Trouser',
    2: 'Pullover',
    3: 'Dress',
    4: 'Coat',
    5: 'Sandal',
    6: 'Shirt',
    7: 'Sneaker',
    8: 'Bag',
    9: 'Ankle boot'
}

plt.figure(figsize=(8, 6))
for i in range(10):
    indices = y_subset == i
    plt.scatter(x_tsne[indices, 0], x_tsne[indices, 1], label=class_labels[i])

plt.title('Visualization of Fashion MNIST')
plt.xlabel(' Dimension 1')
plt.ylabel(' Dimension 2')
plt.legend()
plt.show()
```

```python
data = {'Dimension 1': x_tsne[:, 0], 'Dimension 2': x_tsne[:, 1], 'Class': [class_labels[label] for label in y_subset]}
df = pd.DataFrame(data)
fig = px.scatter(df, x='Dimension 1', y='Dimension 2', color='Class', title='Visualization of Fashion MNIST')
fig.show()
```
